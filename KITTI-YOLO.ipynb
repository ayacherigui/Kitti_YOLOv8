```python
from ultralytics import YOLO
```

```python
#Load the model
model = YOLO('yolov8x.pt')
```

```python
# Training.
results = model.train(
   data='data_configs/data.yaml',
   imgsz=384,
   epochs=100,
   batch=16,
   name='yolov8x_custom')
```
